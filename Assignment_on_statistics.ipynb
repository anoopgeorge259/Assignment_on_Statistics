```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/StudentsPerformance.csv')
```

```python
df.head()
```

```python
# Q1 mean, median, and mode
#math score 	reading score 	writing score

means = df[['math score', 'reading score', 'writing score']].mean()
print("\nMean Values\n" ,means)

medians = df[['math score', 'reading score', 'writing score']].median()
print("\nMedian Values\n" ,medians)

modes = df[['math score', 'reading score', 'writing score']].mode()
print("\n Mode Values\n " ,modes)
```

```python
# Q2 range, variance, and standard deviation

ranges = df[['math score', 'reading score', 'writing score']].max() - df[['math score', 'reading score', 'writing score']].min()
ranges
```

```python
variance = df[['math score', 'reading score', 'writing score']].var()
variance
```

```python
std = df[['math score', 'reading score', 'writing score']].std()
std
```

```python
# Q3 mean, median, and mode of math scores with different parental education levels


df.groupby('parental level of education')['math score'].mean()
```

```python
df.groupby('parental level of education')['math score'].median()
```

```python
df.groupby('parental level of education')['math score'].agg(pd.Series.mode)
```

```python
# Q4 Compare the means of who took the test preparation course versusthose who didn’t.

df.groupby('test preparation course')['math score'].mean()
```

```python
# Q 5  interquartile range (IQR)

print("IQR of Math Score")
print(df['math score'].quantile(0.75) - df['math score'].quantile(0.25))
print("IQR of Reading Score")
print(df['math score'].quantile(0.75) - df['reading score'].quantile(0.25))
print("IQR of Writing Score")
print(df['math score'].quantile(0.75) - df['writing score'].quantile(0.25))
```

```python
#Q6 Std of math score on the basis of type of lunch
df.groupby('lunch')['math score'].std()
```

```python
# Q7 correlation coefficient between math and reading scores, btwn reading and writing scores # Q8

corr = df['math score'].corr(df['reading score'])
print("correlation coefficient between math and reading scores :",corr)
corr = df['reading score'].corr(df['writing score'])
print("correlation coefficient between reading and writing scores :",corr)
#Q8
print("\n Correlation coefficient r > 0 means positive relationship. If the values are closer to 1, that means their correlation strength is very high. \n\
 Therefore here both relation are highly correlated positively. That means both are very closely related. ")
```

```python
# Q9 covariance between math and reading scores.
cov = df['math score'].cov(df['reading score'])
print(cov)

print("Both covariance and correlation are positive relationship,\n\
that means students who score higher in math also probably score higher in reading")
```

```python
# Q10 Z-Test
math_scores = df['math score']

z = (math_scores.mean() - 50) / (math_scores.std(ddof=0) / (len(math_scores) ** 0.5))
float(z)
```

```python
# Q11 T-Test
from scipy import stats

g1 = df[df['lunch']=='free/reduced']['reading score']
g2 = df[df['lunch']=='standard']['reading score']

stats.ttest_ind(g1, g2)
```

```python
# Q12 F-Test
prep = df[df['test preparation course']=='completed']['math score']
no_prep = df[df['test preparation course']=='none']['math score']

f = prep.var(ddof=1) / no_prep.var(ddof=1)

from scipy.stats import f as f_dist
p = 1 - f_dist.cdf(f, len(prep)-1, len(no_prep)-1)

float(f), float(p)
```
